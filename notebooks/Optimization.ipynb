```python
%matplotlib notebook
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return math.sin(float(x)/5.)*math.exp(float(x)/10.)+5*math.exp(-float(x)/2.)
```

```python
minimum = optimize.differential_evolution(f, [(1, 30)])
minimum
```

```python
def h(x):
    return int(f(x))
```

```python
X = np.arange(0, 30, 0.1)
plt.plot(X, np.vectorize(h)(X))
plt.plot(X, np.vectorize(f)(X))
```

```python
minimumH = optimize.differential_evolution(h, [(1, 30)])
minimumH
```
